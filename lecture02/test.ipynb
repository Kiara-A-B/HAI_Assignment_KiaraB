```python
import numpy as np
import pandas as pd

# original data
data = {
    "날짜":["3/20", "3/21", "3/22", "3/23", "3/24"],
    "매출 편차": np.random.randn(5),
    "손님 평균 연령": np.random.randint(10, 60, 5)
}

#create dataframe
df = pd.DataFrame(data)

#print orginal dataframe
print("Original table")
print(df)
print("\n")

#print only 매출 편차 column
print("매출 편차 column")
print(df["매출 편차"])
print("\n")

#20 years ore older
print("20 years or older")
older = df[df["손님 평균 연령"] >=20]
print(older)
print("\n")

#Change values to NaN
print("Change values to NaN")
df.loc[df["날짜"] == "3/22", "매출 편차"] = np.nan
df.iloc[4,df.columns.get_loc("손님 평균 연령")] = np.nan
print(df)
print("\n")

#Check NaN locations
print("Check NaN location")
print(df.isnull())
print("\n")

#Fill 3/22's 매출 편차 with 0
print("Fill 3/22's 매출 편차 with 0")
df['매출 편차'] = df['매출 편차'].fillna(0)
print(df)
print("\n")

#Fill 3/24's 손님 평균 연령 with average
print("#Fill 3/24's 손님 평균 연령 with average")
df['손님 평균 연령'] = df['손님 평균 연령'].fillna(df['손님 평균 연령'].mean())
print(df)
```
